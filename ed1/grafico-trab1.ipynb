```python
# Gráfico Comparativo Completo com Linhas Teóricas de Complexidade
import matplotlib.pyplot as plt
import numpy as np

# Dados dos testes (em segundos)
tamanhos = np.array([10, 100, 1000, 5000])
metodo1 = np.array([0.000000096, 0.000000399, 0.000002633, 0.000008973])  # Inverter Ordem
metodo2 = np.array([0.000000022, 0.000000067, 0.000000398, 0.000000548])  # Busca Sequencial
metodo3 = np.array([0.000000025, 0.000000027, 0.000000029, 0.000000027])  # Busca Binária Iterativa
metodo4 = np.array([0.000000019, 0.000000026, 0.000000038, 0.000000029])  # Busca Binária Recursiva

# Criação das linhas teóricas de complexidade
# Para O(n) - normalizada com base no primeiro ponto do Método 1
fator_on = metodo1[0] / tamanhos[0]
linha_on = fator_on * tamanhos

# Para O(log n) - normalizada com base no primeiro ponto do Método 3
fator_ologn = metodo3[0] / np.log2(tamanhos[0])
linha_ologn = fator_ologn * np.log2(tamanhos)

# Criação do gráfico
plt.figure(figsize=(12, 8))

# Plot dos dados experimentais
plt.plot(tamanhos, metodo1, 'ro-', linewidth=2, markersize=8, label="Método 1 - Inverter Ordem")
plt.plot(tamanhos, metodo2, 'bs-', linewidth=2, markersize=8, label="Método 2 - Busca Sequencial")
plt.plot(tamanhos, metodo3, 'g^-', linewidth=2, markersize=8, label="Método 3 - Busca Binária Iterativa")
plt.plot(tamanhos, metodo4, 'mo-', linewidth=2, markersize=8, label="Método 4 - Busca Binária Recursiva")

# Plot das linhas teóricas
plt.plot(tamanhos, linha_on, 'r--', alpha=0.7, linewidth=2, label="Linha Teórica O(n)")
plt.plot(tamanhos, linha_ologn, 'g--', alpha=0.7, linewidth=2, label="Linha Teórica O(log n)")

# Configurações do gráfico
plt.xlabel("Tamanho do Vetor (n)", fontsize=12)
plt.ylabel("Tempo Médio de Execução (segundos)", fontsize=12)

# Usar escala logarítmica para melhor visualização
plt.xscale('log')
plt.yscale('log')

# Grid e legenda
plt.grid(True, alpha=0.3)
plt.legend(fontsize=10, loc='upper left')

# Ajustar layout
plt.tight_layout()

# Salvar gráfico
plt.savefig("comparacao_complexidade.png", dpi=300, bbox_inches='tight')
plt.show()

# Análise dos resultados
print("=== ANÁLISE DE COMPLEXIDADE ===")
print("\nMétodos O(n) - Crescimento Linear:")
print("• Método 1 (Inverter): Cresce proporcionalmente ao tamanho")
print("• Método 2 (Busca Seq.): Cresce com o tamanho do vetor")

print("\nMétodos O(log n) - Crescimento Logarítmico:")
print("• Método 3 (Busca Bin. Iter.): Tempo quase constante")
print("• Método 4 (Busca Bin. Rec.): Tempo quase constante")

print("\nObservações:")
print("• As buscas binárias são muito mais eficientes que a busca sequencial")
print("• O algoritmo de inversão tem overhead de alocação de memória")
print("• Todos os algoritmos seguem suas complexidades teóricas esperadas")
```

```python
# Gráfico de Barras Comparativo por Tamanho
plt.figure(figsize=(14, 10))

# Configurar dados para gráfico de barras
x = np.arange(len(tamanhos))
width = 0.2

# Criar barras para cada método
plt.bar(x - 1.5*width, metodo1*1000000, width, label='Método 1 - Inverter Ordem', 
        color='red', alpha=0.7)
plt.bar(x - 0.5*width, metodo2*1000000, width, label='Método 2 - Busca Sequencial', 
        color='blue', alpha=0.7)
plt.bar(x + 0.5*width, metodo3*1000000, width, label='Método 3 - Busca Binária Iter.', 
        color='green', alpha=0.7)
plt.bar(x + 1.5*width, metodo4*1000000, width, label='Método 4 - Busca Binária Rec.', 
        color='purple', alpha=0.7)

# Configurações do gráfico
plt.xlabel('Tamanho do Vetor', fontsize=12)
plt.ylabel('Tempo Médio (microssegundos)', fontsize=12)
plt.title('Comparação de Tempos por Tamanho de Vetor\n(em microssegundos)', fontsize=14, fontweight='bold')
plt.xticks(x, tamanhos)
plt.legend()
plt.grid(True, alpha=0.3)

# Adicionar valores nas barras
for i, tamanho in enumerate(tamanhos):
    plt.text(i - 1.5*width, metodo1[i]*1000000 + 0.01, f'{metodo1[i]*1000000:.3f}', 
             ha='center', va='bottom', fontsize=8, rotation=90)
    plt.text(i - 0.5*width, metodo2[i]*1000000 + 0.01, f'{metodo2[i]*1000000:.3f}', 
             ha='center', va='bottom', fontsize=8, rotation=90)
    plt.text(i + 0.5*width, metodo3[i]*1000000 + 0.01, f'{metodo3[i]*1000000:.3f}', 
             ha='center', va='bottom', fontsize=8, rotation=90)
    plt.text(i + 1.5*width, metodo4[i]*1000000 + 0.01, f'{metodo4[i]*1000000:.3f}', 
             ha='center', va='bottom', fontsize=8, rotation=90)

plt.tight_layout()
plt.savefig("comparacao_barras.png", dpi=300, bbox_inches='tight')
plt.show()

# Tabela de resultados
print("\n=== TABELA DE RESULTADOS (microssegundos) ===")
print(f"{'Tamanho':<8} {'Método 1':<10} {'Método 2':<10} {'Método 3':<10} {'Método 4':<10}")
print("-" * 58)
for i, tamanho in enumerate(tamanhos):
    print(f"{tamanho:<8} {metodo1[i]*1000000:<10.3f} {metodo2[i]*1000000:<10.3f} "
          f"{metodo3[i]*1000000:<10.3f} {metodo4[i]*1000000:<10.3f}")

print("\n=== EFICIÊNCIA RELATIVA ===")
print("Comparando com a Busca Sequencial (Método 2):")
for i, tamanho in enumerate(tamanhos):
    if metodo2[i] > 0:
        print(f"Tamanho {tamanho}:")
        print(f"  • Método 3 é {metodo2[i]/metodo3[i]:.1f}x mais rápido")
        print(f"  • Método 4 é {metodo2[i]/metodo4[i]:.1f}x mais rápido")
```
