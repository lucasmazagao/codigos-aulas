```python
import matplotlib.pyplot as plt
import numpy as np

# Dados dos testes (em segundos)
tamanhos = np.array([10, 100, 1000, 5000])
metodo1 = np.array([0.000000096, 0.000000399, 0.000002633, 0.000008973])  # Inverter Ordem
metodo2 = np.array([0.000000022, 0.000000067, 0.000000398, 0.000000548])  # Busca Sequencial
metodo3 = np.array([0.000000025, 0.000000027, 0.000000029, 0.000000027])  # Busca Binria Iterativa
metodo4 = np.array([0.000000019, 0.000000026, 0.000000038, 0.000000029])  # Busca Binria Recursiva

# Criao das linhas tericas de complexidade
# Para O(n) - normalizada com base no primeiro ponto do Mtodo 1
fator_on = metodo1[0] / tamanhos[0]
linha_on = fator_on * tamanhos

# Para O(log n) - normalizada com base no primeiro ponto do Mtodo 3
fator_ologn = metodo3[0] / np.log2(tamanhos[0])
linha_ologn = fator_ologn * np.log2(tamanhos)

# Criao do grfico
plt.figure(figsize=(12, 8))

# Plot dos dados experimentais
plt.plot(tamanhos, metodo1, 'ro-', linewidth=2, markersize=8, label="Mtodo 1 - Inverter Ordem")
plt.plot(tamanhos, metodo2, 'bs-', linewidth=2, markersize=8, label="Mtodo 2 - Busca Sequencial")
plt.plot(tamanhos, metodo3, 'g^-', linewidth=2, markersize=8, label="Mtodo 3 - Busca Binria Iterativa")
plt.plot(tamanhos, metodo4, 'mo-', linewidth=2, markersize=8, label="Mtodo 4 - Busca Binria Recursiva")

# Plot das linhas tericas
plt.plot(tamanhos, linha_on, 'r--', alpha=0.7, linewidth=2, label="Linha Terica O(n)")
plt.plot(tamanhos, linha_ologn, 'g--', alpha=0.7, linewidth=2, label="Linha Terica O(log n)")

# Configuraes do grfico
plt.xlabel("Tamanho do Vetor (n)", fontsize=12)
plt.ylabel("Tempo Mdio de Execuo (segundos)", fontsize=12)

# Usar escala logartmica para melhor visualizao
plt.xscale('log')
plt.yscale('log')

# Grid e legenda
plt.grid(True, alpha=0.3)
plt.legend(fontsize=10, loc='upper left')

# Ajustar layout
plt.tight_layout()

# Salvar grfico
plt.savefig("comparacao_complexidade.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Grfico de Barras Comparativo por Tamanho
plt.figure(figsize=(14, 10))

# Configurar dados para grfico de barras
x = np.arange(len(tamanhos))
width = 0.2

# Criar barras para cada mtodo
plt.bar(x - 1.5*width, metodo1*1000000, width, label='Mtodo 1 - Inverter Ordem', 
        color='red', alpha=0.7)
plt.bar(x - 0.5*width, metodo2*1000000, width, label='Mtodo 2 - Busca Sequencial', 
        color='blue', alpha=0.7)
plt.bar(x + 0.5*width, metodo3*1000000, width, label='Mtodo 3 - Busca Binria Iter.', 
        color='green', alpha=0.7)
plt.bar(x + 1.5*width, metodo4*1000000, width, label='Mtodo 4 - Busca Binria Rec.', 
        color='purple', alpha=0.7)

# Configuraes do grfico
plt.xlabel('Tamanho do Vetor', fontsize=12)
plt.ylabel('Tempo Mdio (microssegundos)', fontsize=12)
plt.xticks(x, tamanhos)
plt.legend()
plt.grid(True, alpha=0.3)

# Adicionar valores nas barras
for i, tamanho in enumerate(tamanhos):
    plt.text(i - 1.5*width, metodo1[i]*1000000 + 0.01, f'{metodo1[i]*1000000:.3f}', 
             ha='center', va='bottom', fontsize=8, rotation=90)
    plt.text(i - 0.5*width, metodo2[i]*1000000 + 0.01, f'{metodo2[i]*1000000:.3f}', 
             ha='center', va='bottom', fontsize=8, rotation=90)
    plt.text(i + 0.5*width, metodo3[i]*1000000 + 0.01, f'{metodo3[i]*1000000:.3f}', 
             ha='center', va='bottom', fontsize=8, rotation=90)
    plt.text(i + 1.5*width, metodo4[i]*1000000 + 0.01, f'{metodo4[i]*1000000:.3f}', 
             ha='center', va='bottom', fontsize=8, rotation=90)

plt.tight_layout()
plt.savefig("comparacao_barras.png", dpi=300, bbox_inches='tight')
plt.show()
```
